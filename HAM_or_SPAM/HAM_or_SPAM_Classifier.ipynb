```python
import string
import nltk
from nltk.corpus import stopwords
```

```python
stopwords.words("english")
```

```python
def text_process(text):
    no_punc=[letter for letter in text if not(letter in string.punctuation)]
    no_punc="".join(no_punc)
    no_punc=no_punc.split()
    return [word for word in no_punc if not(word.lower() in stopwords.words("english"))]
```

```python
text_process("this is a sample message to test the text procesor")
```

```python
messages["message"].head(5).apply(text_process)
```

VECTORIZE!

```python
from sklearn.feature_extraction.text import CountVectorizer
model1=CountVectorizer(analyzer=text_process)
model1.fit(messages["message"])
```

```python
message1=model1.transform(messages["message"])
model1.vocabulary_
```

USE TfidTransformer!

```python
from sklearn.feature_extraction.text import TfidfTransformer

model2=TfidfTransformer()
model2.fit(message1)
```

```python
message2=model2.transform(message1)
model2.idf_[model1.vocabulary_["university"]]
```

USE MultinomialNB!

```python
from sklearn.naive_bayes import MultinomialNB

model3=MultinomialNB()
model3.fit(message2,messages["label"])
```

SPLIT INTO TRAIN AND TEST!

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(messages["message"],messages["label"],test_size=0.3,random_state=101)
```

```python
from sklearn.pipeline import Pipeline

final_model=Pipeline([("countvectorizer",CountVectorizer(analyzer=text_process)),("termfreq",TfidfTransformer()),("multinomial",MultinomialNB())])
final_model.fit(X_train,y_train)
```

PREDICT!

```python
prediction=final_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix

print(classification_report(y_test,prediction))
```
